```python
import os
import string
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from google.colab import files

import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, LSTM, Embedding, Dropout, add
from tensorflow.keras.applications.inception_v3 import InceptionV3, preprocess_input

print("TensorFlow Version:", tf.__version__)

# --- 1. UPLOAD AND UNZIP YOUR DATA ---

print("Please upload the 'My_Flickr_Subset.zip' file you created.")
# This will open an upload dialog
uploaded = files.upload()

if 'My_Flickr_Subset.zip' in uploaded:
    print("Unzipping file...")
    # Unzip the file quietly
    !unzip -q My_Flickr_Subset.zip
    print("Data is ready!")
else:
    print("Error: 'My_Flickr_Subset.zip' not found. Please re-run the cell and upload the correct file.")

# Define file paths
BASE_DIR = 'My_Flickr_Subset'
IMG_DIR = os.path.join(BASE_DIR, 'Images')
CAPTIONS_FILE = os.path.join(BASE_DIR, 'captions.txt')

# --- 2. LOAD AND CLEAN CAPTIONS ---

# Get the list of image filenames you actually uploaded
# This is crucial: we only load captions for these images
uploaded_image_files = os.listdir(IMG_DIR)
uploaded_image_ids = set([filename.split('.')[0] for filename in uploaded_image_files])
print(f"Found {len(uploaded_image_ids)} images in your subset.")

def load_and_clean_captions(captions_filepath, image_ids_subset):
    """
    Loads captions.txt, cleans them, and filters them
    to only include captions for the images in our subset.
    """
    with open(captions_filepath, 'r') as f:
        doc = f.read()

    mapping = {}
    table = str.maketrans('', '', string.punctuation) # To remove punctuation

    for line in doc.split('\n'):
        parts = line.split(',')
        if len(parts) < 2:
            continue

        image_id, caption = parts[0], parts[1:]
        image_id = image_id.split('.')[0] # Remove .jpg

        # ---- This is the most important part ----
        # Only process captions for images we have
        if image_id not in image_ids_subset:
            continue

        # Clean the caption
        caption = " ".join(caption)
        caption = caption.lower()
        caption = caption.translate(table)
        caption = ' '.join(word for word in caption.split() if word.isalpha())
        caption = ' '.join(word for word in caption.split() if len(word) > 1)

        # Add start/end tokens
        caption = 'startseq ' + caption + ' endseq'

        # Store
        if image_id not in mapping:
            mapping[image_id] = []
        mapping[image_id].append(caption)

    return mapping

# Load and process the captions
captions_mapping = load_and_clean_captions(CAPTIONS_FILE, uploaded_image_ids)
print(f"Loaded {len(captions_mapping)} captions for your images.")

# Create a flat list of all captions
all_captions_list = [cap for key in captions_mapping for cap in captions_mapping[key]]

# --- 3. PREPARE TEXT DATA (TOKENIZER) ---

# Create the tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(all_captions_list)

# Get vocabulary size (+1 for the 0-padding)
vocab_size = len(tokenizer.word_index) + 1
print(f"Vocabulary Size: {vocab_size}")

# Get max caption length
max_length = max(len(cap.split()) for cap in all_captions_list)
print(f"Max Caption Length: {max_length}")

# Create word -> index and index -> word mappings
word_to_index = tokenizer.word_index
index_to_word = {i: w for w, i in word_to_index.items()}


# --- 4. PRE-PROCESS IMAGE DATA (INCEPTIONV3) ---

def process_image_features(image_directory, image_ids, target_size=(299, 299)):
    """
    Loads the InceptionV3 model and processes all images in our subset,
    returning a dictionary of {image_id: features}.
    """
    # Load InceptionV3
    base_model = InceptionV3(weights='imagenet')
    # Re-structure model to output the 2048-dim feature vector
    feature_model = Model(inputs=base_model.input, outputs=base_model.layers[-2].output)

    image_features = {}

    for filename in os.listdir(image_directory):
        image_id = filename.split('.')[0]

        # Load and resize the image
        img_path = os.path.join(image_directory, filename)
        img = tf.keras.utils.load_img(img_path, target_size=target_size)

        # Convert to numpy array and preprocess
        x = tf.keras.utils.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = preprocess_input(x)

        # Get features
        feature_vector = feature_model.predict(x, verbose=0)
        image_features[image_id] = feature_vector.reshape((2048,))

    return feature_model, image_features

print("\nProcessing images with InceptionV3...")
# This will take a minute
inception_model, image_features = process_image_features(IMG_DIR, captions_mapping.keys())
print("Image processing complete.")


# --- 5. CREATE DATA SEQUENCES (ALL IN MEMORY) ---
# Since we only have ~100 images, we can load all data into memory
# This is much simpler than a data generator

def create_sequences(tokenizer, max_length, captions_map, features_map, vocab_size):
    X1, X2, y = [], [], [] # X1=features, X2=text, y=next_word

    for image_id, captions_list in captions_map.items():
        # Get the feature vector for this image
        image_feature = features_map[image_id]

        for caption in captions_list:
            # Convert caption to sequence of numbers
            seq = tokenizer.texts_to_sequences([caption])[0]

            # Create N-to-1 sequences
            for i in range(1, len(seq)):
                in_seq, out_seq = seq[:i], seq[i]

                # Pad input sequence
                in_seq = pad_sequences([in_seq], maxlen=max_length)[0]

                # One-hot encode output word
                out_seq = to_categorical([out_seq], num_classes=vocab_size)[0]

                # Append
                X1.append(image_feature)
                X2.append(in_seq)
                y.append(out_seq)

    return np.array(X1), np.array(X2), np.array(y)

print("Creating training sequences...")
X1, X2, y = create_sequences(tokenizer, max_length, captions_mapping, image_features, vocab_size)
print(f"Total sequences created: {len(X1)}")


# --- 6. DEFINE AND TRAIN THE MODEL ---

def define_model(vocab_size, max_length, feature_dim=2048):
    # Image Feature Encoder
    input_features = Input(shape=(feature_dim,))
    fe1 = Dropout(0.5)(input_features)
    fe2 = Dense(256, activation='relu')(fe1)

    # Text Sequence Decoder
    input_seq = Input(shape=(max_length,))
    se1 = Embedding(vocab_size, 256, mask_zero=True)(input_seq)
    se2 = Dropout(0.5)(se1)
    se3 = LSTM(256, use_cudnn=False)(se2) # Added use_cudnn=False to fix the error

    # Merge (Add)
    decoder1 = add([fe2, se3])
    decoder2 = Dense(256, activation='relu')(decoder1)
    outputs = Dense(vocab_size, activation='softmax')(decoder2)

    # Create and compile model
    model = Model(inputs=[input_features, input_seq], outputs=outputs)
    model.compile(loss='categorical_crossentropy', optimizer='adam')

    return model

# Create the model
model = define_model(vocab_size, max_length)
model.summary()

# Train the model
# With ~100 images, this will be very fast
print("\nTraining model...")
model.fit([X1, X2], y, epochs=20, batch_size=64, verbose=2)
print("Model training complete.")


# --- 7. TEST THE MODEL (GENERATE CAPTIONS) ---

def generate_caption_greedy(model, tokenizer, image_feature, max_length, index_to_word_map):
    """
    Generates a caption for an image using a greedy search.
    """
    in_text = 'startseq'

    for i in range(max_length):
        # Convert current text to a sequence
        sequence = tokenizer.texts_to_sequences([in_text])[0]
        # Pad the sequence
        sequence = pad_sequences([sequence], maxlen=max_length)

        # Predict the next word
        # Note: We need to provide the image feature as a batch (np.array([image_feature]))
        yhat = model.predict([np.array([image_feature[0]]), sequence], verbose=0)

        # Get the word with the highest probability
        word_index = np.argmax(yhat)
        word = index_to_word_map.get(word_index)

        # Stop if we can't find the word or if it's 'endseq'
        if word is None:
            break

        in_text += ' ' + word

        if word == 'endseq':
            break

    # Clean up the final caption
    final_caption = in_text.split(' ', 1)[1] # Remove 'startseq'
    final_caption = final_caption.rsplit(' ', 1)[0] # Remove 'endseq'
    return final_caption

# --- Let's test it! ---
print("\n--- TESTING THE MODEL ---")

# Get one of the image IDs you uploaded
test_image_id = list(captions_mapping.keys())[0]
test_image_path = os.path.join(IMG_DIR, test_image_id + '.jpg')

# Get the pre-computed feature
test_image_feature = image_features[test_image_id]
# Reshape feature for model (it needs a batch dimension)
test_image_feature_batch = np.array([test_image_feature])

# Generate a caption
predicted_caption = generate_caption_greedy(
    model,
    tokenizer,
    test_image_feature_batch,
    max_length,
    index_to_word
)

# Show the image
img = Image.open(test_image_path)
plt.imshow(img)
plt.axis('off')
plt.show()

# Print results
print(f"IMAGE: {test_image_path}")
print("\n--- PREDICTED CAPTION ---")
print(predicted_caption)

print("\n--- ACTUAL CAPTIONS ---")
for cap in captions_mapping[test_image_id]:
    # Clean 'startseq' and 'endseq' for printing
    clean_cap = cap.replace('startseq ', '').replace(' endseq', '')
    print(clean_cap)
```
